# PCA Test

## About PCA

PCA — is a method to minimize data features shape and make a visual representation of classification signs

If you have a lot of feature-columns and can’t imagine if the algorithm will be able to give a qualitative breakdown of the classification — PCA is a good choice

## Import modules

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import mglearn
```

## Load data and scaling

```python
cancer = load_breast_cancer()
scaler = StandardScaler().fit(cancer.data)
X_scaled = scaler.transform(cancer.data)
```

## PCA

```python
pca = PCA(n_components=2)
pca.fit(X_scaled)
```

## PCA shape result 

```python
X_pca = pca.transform(X_scaled)
print(X_scaled.shape)
print(X_pca.shape)
```

## PCA plot result

```python
plt.figure(figsize=(10,10))
mglearn.discrete_scatter(X_pca[:, 0], X_pca[:, 1], cancer.target)
plt.legend(cancer.target_names, loc="best")
plt.gca().set_aspect("equal")
plt.xlabel("First main component")
plt.ylabel("Second main component")
plt.show()
```

## Components

```python
print('First component:\n{}\n'.format(pca.components_[0]))
print('Second component:\n{}'.format(pca.components_[1]))
```
