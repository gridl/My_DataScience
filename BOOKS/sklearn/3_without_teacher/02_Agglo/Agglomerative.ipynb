# Agglomerative Clustering

## About

In start of clustering process this algorythm make every dot of data like a separated cluster

And unite the nearest dots until get n_clusters = \*your choise\*

## mglearn example

```python
import mglearn
mglearn.plots.plot_agglomerative_algorithm()
```

## Import

```python
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn.cluster import AgglomerativeClustering
```

## BLOBS

```python
X, Y = make_blobs(random_state=40)
mglearn.discrete_scatter(X[:, 0], X[:, 1])
```

### n_clusters = 2

GIVE IT MORE!

```python
agg = AgglomerativeClustering(n_clusters=2)
assigment = agg.fit_predict(X)
mglearn.discrete_scatter(X[:, 0], X[:, 1], assigment)
```

### n_clusters = 3
