# KMeans

## About algorithm

Simple and usefull algorithm of clustering

He random-generates centres of clusters and calculate distantion between dots

Steps of his actions:

- Take a inner data
- Initialization
- =LOOP=
- - Make centres
- - Recalculate centres
- - Recalcelate clusters
- =FINAL=

### Example of KMeans jop

```python
import mglearn
mglearn.plots.plot_kmeans_algorithm()
```

## Imports

```python
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
```

## BLOBS

### w/o clustering

```python
X, Y = make_blobs(random_state=40)
mglearn.discrete_scatter(X[:, 0], X[:, 1], markers='o')
```

### n_clusters = 2

Not so good

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
mglearn.discrete_scatter(X[:, 0], X[:, 1], kmeans.labels_, markers='o')
```

### n_clusters = 3

Perfect

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
mglearn.discrete_scatter(X[:, 0], X[:, 1], kmeans.labels_, markers='o')
```

### n_clusters = 5

Too much

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
mglearn.discrete_scatter(X[:, 0], X[:, 1], kmeans.labels_, markers='o')
```

### n_clusters = 8

PLS STOP IT!

```python
kmeans = KMeans(n_clusters=8)
kmeans.fit(X)
mglearn.discrete_scatter(X[:, 0], X[:, 1], kmeans.labels_, markers='o')
```
