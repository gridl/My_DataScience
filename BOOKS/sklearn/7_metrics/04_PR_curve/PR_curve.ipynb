# PR_curve

## Imports

```python
import numpy as np

from sklearn.metrics import precision_recall_curve

from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

import matplotlib.pyplot as plt
%matplotlib inline
```

# Load data and PLOT

```python
X, Y = make_blobs(n_samples=4000, centers=2, cluster_std=5, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=42)

plt.figure(figsize=(15,15))
plt.plot(X[:,0], X[:,1], 'ro')
```

## LogReg

```python
model = LogisticRegression()
model.fit(X_train, y_train)
model_defu = model.decision_function(X_test)
prec, rec, thr = precision_recall_curve(y_test, model_defu)
close_zero = np.argmin(np.abs(thr))

plt.figure(figsize=(15,15))

plt.plot(prec[close_zero], rec[close_zero], 'ko', markersize='15', fillstyle='none', mew=2, label='close_zero_dot')
plt.plot(prec, rec, 'r-', label='prec_rec_curve')

plt.legend(loc=3, prop={'size':27})
plt.title('precision  recall', fontsize=27)
plt.xlabel('precision', fontsize=20)
plt.ylabel('recall', fontsize=20)
```

## SVC

```python
model = SVC()
model.fit(X_train, y_train)
model_defu = model.decision_function(X_test)
prec, rec, thr = precision_recall_curve(y_test, model_defu)
close_zero = np.argmin(np.abs(thr))

plt.figure(figsize=(15,15))

plt.plot(prec[close_zero], rec[close_zero], 'ko', markersize='15', fillstyle='none', mew=2, label='close_zero_dot')
plt.plot(prec, rec, 'r-', label='prec_rec_curve')

plt.legend(loc=3, prop={'size':27})
plt.title('precision  recall', fontsize=27)
plt.xlabel('precision', fontsize=20)
plt.ylabel('recall', fontsize=20)
```
