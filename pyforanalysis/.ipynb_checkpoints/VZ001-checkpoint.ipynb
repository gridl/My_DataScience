```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

plt.plot(np.random.randn(50).cumsum(), 'k--')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1,)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

plt.plot(np.random.randn(50).cumsum(), 'k--')
_ = ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(20), np.arange(20) + 3 * np.arange(20))
```

### subplots_adjust

#### w/o

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
```

#### w

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500), bins=25, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

## Colors, markers, line styles

#### one way

```python
plt.plot(np.random.randn(30).cumsum(), 'ko--')
```

#### another way

```python
plt.plot(np.random.randn(30).cumsum(), color='k', marker='o', linestyle='--')
```

#### bonus: another interpolation

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

## Ticks and another plot decorators

```python
plt.plot(data, 'k--', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
plt.xlim()
```

```python
plt.plot(data, 'k--', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
plt.xlim([0,10])
```

## Labels

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())

ticks = ax.set_xticks([0, 250, 500, 750, 1000])
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())

ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], 
                            rotation=30, 
                            fontsize='small')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())

ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], 
                            rotation=30, 
                            fontsize='small')
ax.set_title('My plot')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())

ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], 
                            rotation=30, 
                            fontsize='small')
ax.set_title('My plot')
ax.set_xlabel('Stages')
```

## Labels

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'b--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'r.', label='three')
ax.plot(np.random.randn(1000).cumsum(), 'g^', label='_nolegend_')
ax.legend(loc='best', fontsize=20)
```
