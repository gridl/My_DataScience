# Simple Linear Regression with synthetic data

This Data I made in EXCEL

## True formula:

### → Y = X + 55

## OR

### → Y = X + 50 + randbetween(0;10)

```python
import pandas as pd

data = pd.read_csv('synthetic_data.csv', sep=';')

orig_i = 55
orig_c = 1

data.head(10)
```

```python
data.describe()
```

```python
data.corr()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline

# ==================
# ORIGINAL DOTES PLOT
# ==================

figure(figsize=(10,10))
plt.ylim(50, 110)
plt.plot(data['X'], data['Y'], 'go', markersize=8, label='DataSet')
plt.legend(loc=2, prop={'size': 20})
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array(data['X']).reshape(-1,1)
Y = np.array(data['Y']).reshape(-1,1)

lr = LinearRegression()
lr.fit(X,Y)
```

```python
print('FORMULA:')
print('Y = c*X + i', end='\n\n')

print('c: {:.2f}\t\t → 1'.format(lr.coef_[0][0]))
print('i: {:.2f}\t → 55'.format(lr.intercept_[0]))
print('R^2: {:.2f}'.format(lr.score(X,Y)))
```

```python
ax = np.linspace(1,50,50, dtype='int')         # make axis
pred = ax * lr.coef_[0][0] + lr.intercept_[0]  # make prediction line
orig = ax * orig_c + orig_i                    # make original line
```

```python
# ===========================
# ORIGINAL DOTES + PREDICTION
# ===========================

figure(figsize=(10,10))
plt.ylim(50, 110)
plt.plot(ax, pred, 'b-', linewidth=4, label='Prediction')
plt.plot(data['X'], data['Y'], 'go', markersize=8, label='DataSet')
plt.legend(loc=2, prop={'size': 20})
```

```python
# ===========================================
# ORIGINAL DOTES + PREDICTION + ORIGINAL LINE
# ===========================================

figure(figsize=(10,10))
plt.ylim(50, 110)
plt.plot(ax, pred, 'b-', linewidth=4, label='Prediction')
plt.plot(data['X'], data['Y'], 'go', markersize=8, label='DataSet')
plt.plot(ax, orig, 'm--', linewidth=4, label='Origin Line')
plt.legend(loc=2, prop={'size': 20})
```
