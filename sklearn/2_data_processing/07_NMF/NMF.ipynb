# NMF
Non-negative Matrix Factorization

## About

This method find more additive features and minimize shapes of matrix

## mglearn presentation

```python
import mglearn
mglearn.plots.plot_nmf_illustration()
```

## Imports

```python
import pandas as pd

from sklearn.datasets import load_boston
from sklearn.decomposition import NMF

import matplotlib.pyplot as plt
%matplotlib inline
```

# DataSet 

head(10)

```python
data = load_boston()
data = pd.DataFrame(data.data, columns=data.feature_names)
data.head(10)
```

## Description

```python
desc = data.describe()
desc.loc['min']
```

## Realization NMF with two components

```python
nmf = NMF(n_components=2, random_state=0)
nmf.fit(data)

data_nmf = nmf.transform(data)
data_nmf = pd.DataFrame(data_nmf)

plt.figure(figsize=(12,12))
plt.plot(data_nmf.loc[:,0], data_nmf.loc[:,1], 'ro')
```
