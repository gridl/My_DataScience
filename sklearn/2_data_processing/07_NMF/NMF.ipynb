```python
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.decomposition import NMF
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = load_boston()
data = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
desc = data.describe()
data.head(10)
```

```python
desc.loc['min']
```

```python
nmf = NMF(n_components=2, random_state=0)
nmf.fit(data)
```

```python
data_nmf = nmf.transform(data)
data_nmf = pd.DataFrame(data_nmf)
```

```python
data_nmf.head(10)
```

```python
plt.figure(figsize=(12,12))
plt.plot(data_nmf.loc[:,0], data_nmf.loc[:,1], 'ro')
```
