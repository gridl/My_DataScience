# DBSCAN

## About

DBSCAN is a powerful algorythm of clustering

He has a more sensitivity params, but can make beauty result

## How it works

In a giant hall a crowd of people celebrates someone's birthday

Someone loiter one, but most - with friends

Some companies are simply crowded with a crowd, some - dance in circles or dance lambadas

We want to divide the people in the hall into groups

**PARAMS**

- min_samples — is a minimum people in group, wich we can imagine as cluster
- eps — closenes of people (for exmpl: length of hand)



## mglearn representation

```python
import mglearn
mglearn.plots.plot_dbscan()
```

## Imports

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
```

## BLOBS

```python
X, Y = make_blobs(random_state=40, n_samples=100)
mglearn.discrete_scatter(X[:,0], X[:,1], markers='o')
```

### Nude DBScan

Pah!

```python
dbscan = DBSCAN()
fp = dbscan.fit_predict(X)
mglearn.discrete_scatter(X[:,0], X[:,1], fp, markers='o')
```

### eps = 1

Good!

```python
dbscan = DBSCAN(eps=1)
fp = dbscan.fit_predict(X)
mglearn.discrete_scatter(X[:,0], X[:,1], fp, markers='o')
```

### eps = 2

Best

```python
dbscan = DBSCAN(eps=2)
fp = dbscan.fit_predict(X)
mglearn.discrete_scatter(X[:,0], X[:,1], fp, markers='o')
```

## MOONS

```python
X, Y = make_moons(n_samples=777, noise=0.05)
mglearn.discrete_scatter(X[:,0], X[:,1], markers='o')
```

### Nude DBScan

no changes -__-

```python
dbscan = DBSCAN()
fp = dbscan.fit_predict(X)
mglearn.discrete_scatter(X[:,0], X[:,1], fp, markers='o')
```

### eps = 0.1

Short hands can make magic \*\_\_\_\_\_\*

```python
dbscan = DBSCAN(eps=0.1)
fp = dbscan.fit_predict(X)
mglearn.discrete_scatter(X[:,0], X[:,1], fp, markers='o')
```
