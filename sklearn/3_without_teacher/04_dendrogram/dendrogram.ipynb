# Dendrogram 

with aglomerative cluster kenrel

## Imports

```python
from scipy.cluster.hierarchy import dendrogram, ward
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

## BLOBS

```python
X, y = make_blobs(random_state=40, n_samples=54)

X_la = ward(X)

plt.figure(figsize=(15,15))
dendrogram(X_la, show_leaf_counts=False)
plt.show()
```

## MOONS

```python
X, y = make_moons(random_state=40, n_samples=54)

X_la = ward(X)

plt.figure(figsize=(15,15))
dendrogram(X_la, show_leaf_counts=False)
plt.show()
```
