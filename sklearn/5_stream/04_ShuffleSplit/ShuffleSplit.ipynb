## Imp

```python
from sklearn.datasets import load_iris

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import cross_val_score, ShuffleSplit

import numpy as np
```

## matplotlib settings

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

font = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : 17}

matplotlib.rc('font', **font)
```

## Data

```python
iris = load_iris()

logreg = LogisticRegression()
knc = KNeighborsClassifier()
```

## splits: 3

```python
shsp = ShuffleSplit(n_splits=3, random_state=0)

scores = cross_val_score(logreg, iris.data, iris.target, cv=shsp)
print('= LOGREG =')
print('Scores:\n{}\n'.format(scores))
print('Mean:\n{:.3f}\n'.format(scores.mean()))

scores = cross_val_score(knc, iris.data, iris.target, cv=shsp)
print('= KNC =')
print('Scores:\n{}\n'.format(scores))
print('Mean:\n{:.3f}\n'.format(scores.mean()))
```

## splits: 5

```python
shsp = ShuffleSplit(n_splits=5, random_state=0)

scores = cross_val_score(logreg, iris.data, iris.target, cv=shsp)
print('= LOGREG =')
print('Scores:\n{}\n'.format(scores))
print('Mean:\n{:.3f}\n'.format(scores.mean()))

scores = cross_val_score(knc, iris.data, iris.target, cv=shsp)
print('= KNC =')
print('Scores:\n{}\n'.format(scores))
print('Mean:\n{:.3f}\n'.format(scores.mean()))
```

## splits: 8

```python
shsp = ShuffleSplit(n_splits=8, random_state=0)

scores = cross_val_score(logreg, iris.data, iris.target, cv=shsp)
print('= LOGREG =')
print('Scores:\n{}\n'.format(scores))
print('Mean:\n{:.3f}\n'.format(scores.mean()))

scores = cross_val_score(knc, iris.data, iris.target, cv=shsp)
print('= KNC =')
print('Scores:\n{}\n'.format(scores))
print('Mean:\n{:.3f}\n'.format(scores.mean()))
```

## splits: 13

```python
shsp = ShuffleSplit(n_splits=13, random_state=0)

scores = cross_val_score(logreg, iris.data, iris.target, cv=shsp)
print('= LOGREG =')
print('Scores:\n{}\n'.format(scores))
print('Mean:\n{:.3f}\n'.format(scores.mean()))

scores = cross_val_score(knc, iris.data, iris.target, cv=shsp)
print('= KNC =')
print('Scores:\n{}\n'.format(scores))
print('Mean:\n{:.3f}\n'.format(scores.mean()))
```

## LOOP: n_splits â†’ [1: 200]

```python
logreg_scores = []
knc_scores = []

for ns in range(1, 201):
    shsp = ShuffleSplit(n_splits=ns, random_state=0)
    
    scores = cross_val_score(logreg, iris.data, iris.target, cv=shsp)
    logreg_scores.append(scores.mean())
    
    scores = cross_val_score(knc, iris.data, iris.target, cv=shsp)
    knc_scores.append(scores.mean())
```

## Plot

```python
plt.figure(figsize=(15,15))

plt.plot(logreg_scores, 'r-', label='LogReg')
plt.plot(knc_scores, 'b-', label='KNC')

plt.ylim(bottom=0.95, top=1.01)

plt.title('loop: score(n_splits)')
plt.xlabel('n_splits - 1')
plt.ylabel('score')

plt.legend(loc=1)
plt.grid()

plt.show()
```
